```python
import requests
import os
os.mkdir("albums")
res = requests.get("https://www.billboard.com/charts/year-end/top-billboard-200-albums/")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(res.text, 'html.parser')
```

```python
images = soup.select(".c-lazy-image__img.lrv-u-background-color-grey-lightest.lrv-u-width-100p.lrv-u-display-block.lrv-u-height-auto")
print(len(images))
print(images)
```

```python
image_number = 0
for i in images:
    link = images[i]['data-lazy-src']
    print(link)
        
    image_link = requests.get(link)
    
    f = open('albums/myImage' + str(image_number) + '.png', 'wb')
    image_number = image_number + 1
    f.write(image_link.content)
    f.close()

print(str(image_number) + ' out of ' + str(len(images)) + ' were saved successfully')
```
