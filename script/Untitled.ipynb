```python
import requests
import os

res = requests.get("https://animebee.to/best-anime-2022")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(res.text, 'html.parser')
print(soup)
```

```python
images = soup.select("img.film-poster-img")
print(len(images))
print(images)
```

```python
image_number = 0
for i in images:
    link = i['data-src']
    print(link)
        
    image_link = requests.get(link)
    
    f = open('anime/myImage' + str(image_number) + '.png', 'wb')
    image_number = image_number + 1
    f.write(image_link.content)
    f.close()
    if image_number > 10:
        break

print(str(image_number) + ' out of ' + str(len(images)) + ' were saved successfully')
```
